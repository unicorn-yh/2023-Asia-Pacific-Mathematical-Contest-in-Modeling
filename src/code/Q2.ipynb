### 1. Import Libraries

```python
import json
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
from sklearn.cluster import KMeans
```

```python
json_path = "../Result/Counting/json/"
img_path = "../Attachment/Attachment 1/"
```

### 2. Plot Scatter Diagram

```python
def question2(json_file):
    ans = []
    with open(json_file,'r') as file:
        data_label = json.load(file)
    for circle in data_label:
        point1, point2 = circle['point']
        radius = circle['radius']
        ans.append((point1,point2))
    return ans

x = []
y = []
for i in range(1,201):
    json_file = json_path + str(i) + ".json"
    points_list = question2(json_file)
    for j in points_list:
        x.append(j[0])
        y.append(j[1])

plt.figure(figsize=(8,8))
plt.title("The geometric coordinates of all apples")
plt.scatter(x,y)
plt.ylabel(". ")
plt.xlabel("x-axis")

plt.tight_layout()
plt.show()
```
